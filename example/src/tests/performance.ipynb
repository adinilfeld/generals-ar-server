```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300


def plot(tasks_per_worker):
    push_average_times_col = []
    pull_average_times_col = []
    local_average_times_col = []
    worker_counts = []
    
    for i in range(1, 9):
        worker_counts.append(i)
        # Push
        df = pd.read_csv(f'../results/push_{i}.csv')
        avg_time_col = df[str(tasks_per_worker)].mean()
        push_average_times_col.append(avg_time_col)
        # Pull
        df = pd.read_csv(f'../results/pull_{i}.csv')
        avg_time_col = df[str(tasks_per_worker)].mean()
        pull_average_times_col.append(avg_time_col)
        # Local
        df = pd.read_csv(f'../results/local_{i}.csv')
        avg_time_col = df[str(tasks_per_worker)].mean()
        local_average_times_col.append(avg_time_col)

    plt.errorbar(worker_counts, push_average_times_col, label='Push', yerr=1.0, capsize=5, marker='o')
    plt.errorbar(worker_counts, pull_average_times_col, label='Pull', yerr=1.0, capsize=5, marker='o')
    plt.errorbar(worker_counts, local_average_times_col, label='Local', yerr=1.0, capsize=5, marker='o')
    plt.xlabel('Number of Workers')
    plt.ylabel('Average Completion Time')
    plt.ylim(bottom=0)
    plt.title(f'Average Completion Time vs. Number of Workers\nTasks per Worker = {tasks_per_worker}')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'../../figures/average_time_plot_{tasks_per_worker}.png', dpi=300)
```

```python
plot(1)
```

```python
plot(3)
```
